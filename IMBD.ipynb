```python
#Notebook Final: Analisis Sentimen IMDB dengan ANN
# =========================================
# 1. Load Data & EDA Singkat
# =========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset IMDB
data = pd.read_csv("IMDB Dataset.csv")  # Pastikan file ada di folder kerja
data.head()

# Info dataset
data.info()

# Distribusi target
plt.figure(figsize=(6,4))
sns.countplot(x='sentiment', data=data, palette=['#EED4E5','#C2C4E2'])
plt.title("Distribusi Sentimen (Positif vs Negatif)")
plt.xlabel("Sentimen")
plt.ylabel("Jumlah")
plt.show()

# Contoh ulasan
print("Contoh Review:\n", data['review'][0])
print("Label Sentimen:", data['sentiment'][0])

# =========================================
# 2. Preprocessing (Encoding & Scaling)
# =========================================
import re
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Fungsi pembersihan teks
def clean_text(text):
    text = text.lower()                       # lowercase
    text = re.sub(r"<.*?>", " ", text)       # hapus HTML tags
    text = re.sub(r"[^a-zA-Z]", " ", text)   # hapus karakter non-letters
    text = re.sub(r"\s+", " ", text)         # hapus spasi ekstra
    return text

# Terapkan pembersihan teks
data['review_clean'] = data['review'].apply(clean_text)

# Label encoding target (pos=1, neg=0)
label_encoder = LabelEncoder()
data['sentiment_label'] = label_encoder.fit_transform(data['sentiment'])

# TF-IDF vektorisasi teks (maks 5000 fitur)
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(data['review_clean']).toarray()
y = data['sentiment_label'].values

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# =========================================
# 3. Definisi Model (Arsitektur ANN)
# =========================================
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

# Tentukan jumlah input
input_dim = X_train.shape[1]

# Bangun model Sequential
model = Sequential([
    Dense(128, activation='relu', input_dim=input_dim),
    Dense(64, activation='relu'),
    Dropout(0.3),          # Regularisasi
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # Output biner
])

# Ringkasan model
model.summary()

# =========================================
# 4. Training Process
# =========================================
from keras import callbacks

# Compile model
opt = Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

# Early stopping untuk mencegah overfitting
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss', 
    patience=5, 
    restore_best_weights=True
)

# Latih model
history = model.fit(
    X_train, y_train,
    validation_split=0.2,  # 20% dari train sebagai validation
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping]
)

# =========================================
# 5. Visualisasi Grafik (Loss & Accuracy)
# =========================================
import pandas as pd

# Convert history ke dataframe
history_df = pd.DataFrame(history.history)

# Plot Loss
plt.figure(figsize=(8,5))
plt.plot(history_df['loss'], label='Training Loss', color='#BDE2E2')
plt.plot(history_df['val_loss'], label='Validation Loss', color='#C2C4E2')
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Plot Accuracy
plt.figure(figsize=(8,5))
plt.plot(history_df['accuracy'], label='Training Accuracy', color='#BDE2E2')
plt.plot(history_df['val_accuracy'], label='Validation Accuracy', color='#C2C4E2')
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# =========================================
# 6. Evaluasi Akhir
# =========================================
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Prediksi test set
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Confusion Matrix
cf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negatif','Positif'], yticklabels=['Negatif','Positif'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - IMDB Sentiment")
plt.show()

# Classification Report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Untuk ANN
model.save("sentiment_ann.h5")

# Untuk Naive Bayes
import joblib
joblib.dump(model, "sentiment_nb.pkl")
joblib.dump(tfidf, "tfidf_vectorizer.pkl")  # simpan TF-IDF vectorizer
```
